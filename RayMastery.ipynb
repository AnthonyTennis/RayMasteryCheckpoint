```python
%load_ext autoreload
%autoreload 2
import numpy as np
import sys
import ray
import SampleSort
import qs
ray.init()
```

```python
def is_sorted(arr):
    if len(arr) <= 1:
        return True
    for i in range(len(arr) - 2):
        if arr[i] > arr[i+1]:
            return False
    return True
```

```python
# test this function
arr_1 = []          # t
arr_2 = [1]         # t
arr_3 = [1,1]       # t
arr_4 = [1,2,3,4,5] # t
arr_5 = [1,3,5,2,1] # f
results = [is_sorted(arr_1),is_sorted(arr_2),is_sorted(arr_3),is_sorted(arr_4),is_sorted(arr_5)]
results
```
